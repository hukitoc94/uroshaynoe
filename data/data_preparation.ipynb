### Мокрое просеевание 
#### Урожайное 


```python
df = pd.read_excel('row_data/урожайное_мокрое в работу.xlsx')
df["GPS №"]  = df['Варианты опыта'].str.split(' ', expand=True)[0]
df["Год"]  = df['Варианты опыта'].str.split(' ', expand=True)[1]
df["Тип обработки"]  = df['Варианты опыта'].str.split(' ', expand=True)[2]
df["повторность"]  = df['Варианты опыта'].str.split(' ', expand=True)[4]
df = df.iloc[:,1:]


mokroe = df.copy()

mokroe['Вес образца, г'] = mokroe["вес чашки+просеянная фракция агрегатов"] - mokroe["вес пустой чашки в которую просеивали"]
mokroe = mokroe.drop(["Масса образца", "вес пустой чашки в которую просеивали","вес чашки+просеянная фракция агрегатов" ], axis = 1)
mokroe = mokroe.reindex(columns = ['Год', "Тип обработки","GPS №", "повторность",  "Размер сухого агрегата", "Размеры фракций", 'Вес образца, г'])

mokroe["Размер сухого агрегата"] = mokroe["Размер сухого агрегата"].str.replace('мм', '')
mokroe["Размер сухого агрегата"] = mokroe["Размер сухого агрегата"].str.replace(' ', '')

mokroe["Размеры фракций"] = mokroe["Размеры фракций"].str.replace('мм', '')
mokroe["Размеры фракций"] = mokroe["Размеры фракций"].str.replace(' ', '')


sum_ = mokroe.groupby(["Год","Тип обработки", "GPS №", "повторность", "Размер сухого агрегата"	]).sum().reset_index()


sum_['потери'] = 20 - sum_["Вес образца, г"]


sum_["Размеры фракций"] = "<0.25"
sum_ = sum_.drop(['Вес образца, г'], axis = 1)


mokroe = pd.merge(sum_,mokroe,how = 'outer' , on = ['Год',"Тип обработки",'повторность', "GPS №", 'Размер сухого агрегата', 'Размеры фракций'] )

mokroe = mokroe.drop("потери", axis = 1)
mokroe['region'] = "урожайное"
mokroe.to_csv('мокрое просеивание урожайное.csv', index = False)

del mokroe, sum_ ,df 
mokroe = pd.read_csv('мокрое просеивание урожайное.csv')
mokroe
```

### считаем все агрофиз показатели

Расчет агрофизических параметров по сухому просеиванию -

- СВД мм
- СВД агроном ценных
- Эрозионно опасные <1 мм



Расчет агрофизических параметров по мокрому просеиванию

- СВД по фракциям >10 , 10-7 , 7-5, 5-3
- Эрозионно опасные <1 мм по фракциям >10 , 10-7 , 7-5, 5-3


надо трансформировать голые данные по сухому и мокрому просеиванию в готовые к расчетам и визуализации данным

#### пока делаем все по курску потом продублируем для ставрополя 

```python
big_suhoe
#подготовка размеров воздушносухих агрегатов
big_suhoe['size'] = big_suhoe["Размер агрегатов, мм"]
big_suhoe['size'] = big_suhoe['size'].str.replace('>','16-')
big_suhoe['size'] = big_suhoe['size'].str.replace('<0.25','0.25-0')
big_suhoe[['bottom' , 'top']] = big_suhoe['size'].str.split("-", expand = True)
big_suhoe[['bottom' , 'top']] = big_suhoe[['bottom' , 'top']].astype(float)
big_suhoe['size'] = (big_suhoe['top'] + big_suhoe['bottom']) / 2 

#свд воздушносухие
big_suhoe['СВД, мм'] = (big_suhoe["Содержание агрегатов, %"] * big_suhoe['size'])/100
SVD = big_suhoe.groupby(["region","Год",'Тип обработки', "GPS №","Поврторность"]).sum().reset_index()[["region","Год",'Тип обработки', "GPS №","Поврторность","СВД, мм"]]

#свд воздушносухие агрономически ценные
agr_cen = big_suhoe[~big_suhoe["Размер агрегатов, мм"].isin(['>10','<0.25'])]
agr_cen['СВД агроном. цен., мм'] = (agr_cen["Содержание агрегатов, %"] * agr_cen['size'])/100
agr_cen = agr_cen.groupby(["region","Год",'Тип обработки', "GPS №","Поврторность"]).sum().reset_index()[["region","Год",'Тип обработки', "GPS №","Поврторность",'СВД агроном. цен., мм']]
#Эрозионно опасные <1 мм
eroz = big_suhoe[big_suhoe["Размер агрегатов, мм"].isin(['1-0.5', '0.5-0.25','<0.25'])]
eroz = eroz.groupby(["region","Год",'Тип обработки', "GPS №","Поврторность"]).sum().reset_index()[["region","Год",'Тип обработки', "GPS №","Поврторность","Содержание агрегатов, %"]]

agrofiz = SVD.copy()
agrofiz['СВД агроном. цен., мм'] = agr_cen['СВД агроном. цен., мм']
agrofiz['Эрозионно опасные (<1 мм), %'] = eroz["Содержание агрегатов, %"]
#агрофизика по воздушносухим 
agrofiz.head(5)
```

```python
big_mokroe['size'] = big_mokroe["Размеры фракций"]
big_mokroe['size'] = big_mokroe['size'].str.replace('>10','16-10')
big_mokroe['size'] = big_mokroe['size'].str.replace('>7','10-7')
big_mokroe['size'] = big_mokroe['size'].str.replace('>5','5-4')
big_mokroe['size'] = big_mokroe['size'].str.replace('>3','4-3')
big_mokroe['size'] = big_mokroe['size'].str.replace('<0.25','0.25-0')
big_mokroe[['bottom' , 'top']] = big_mokroe['size'].str.split("-", expand = True)
big_mokroe[['bottom' , 'top']] = big_mokroe[['bottom' , 'top']].astype(float)
big_mokroe['size'] = (big_mokroe['top'] + big_mokroe['bottom']) / 2 

big_mokroe['СВД, мм'] = big_mokroe["Содержание агрегатов, %"] * big_mokroe['size']/100

svd_mokroe = big_mokroe.groupby(["region","Год",'Тип обработки', "GPS №","повторность"]).sum().reset_index()

for size in big_mokroe["Размер сухого агрегата"].unique():
    svd_mokroe["СВД водоустойчивые {}, мм".format(size)] = big_mokroe[big_mokroe["Размер сухого агрегата"] == size].groupby(["region","Год",'Тип обработки', "GPS №","повторность"]).sum().reset_index()["СВД, мм"]


svd_mokroe = svd_mokroe[["region",
                            "Год",
                            'Тип обработки',
                             "GPS №",
                             "повторность",
                             "СВД водоустойчивые 10-7, мм",
                             "СВД водоустойчивые 5-3, мм",
                             "СВД водоустойчивые 7-5, мм",
                             "СВД водоустойчивые >10, мм"]]

agrofiz['повторность'] = agrofiz['Поврторность'] 
agrofiz = pd.merge(agrofiz, svd_mokroe,how = 'outer' , on = ["region","Год",'Тип обработки', "GPS №","повторность"] )

agrofiz.head(10)
```
